### header(열 이름) 설정 

````python
# header = 0 
# header = None : 첫번째 행이 컬럼명이 아님 

excel_header = pd.read_excel('rawData_sheetplus.xlsx',header=0)
excel_header
````

````python
# 헤더 변경
# naes =[''] : 컬럼 수 와 동일해야함
excel_header = pd.read_excel('rawData_sheetplus.xlsx', names=['연차','연도','매출','순이익','직원수'])
excel_header
````

````python
# 헤더를 설정하고 싶지 안하을 때 -> header = None 
excel_header = pd.read_excel('rawData_sheetplus.xlsx',header=None, names=['연차','연도','매출','순이익','직원수'])
excel_header
````

### 인덱스 설정 

````python
# incex_col='' : 인덱스 설정 
excel_index = pd.read_excel('rawData_sheetplus.xlsx',index_col='연차')
excel_index
````

````python
# incex_col='' : 인덱스 없이 불러올 때 ( 불러온 이후 인덱스 지정)
excel_index = pd.read_excel('rawData_sheetplus.xlsx',index_col=None)
excel_index
````

### 사용할 열 만 선택

````python
# usecols=['']
# 원하는 컬럼만 불러오고싶을 때 

excel_index = pd.read_excel('rawData_sheetplus.xlsx',usecols=['연차','매출'])
excel_index
````

### 제한할 행 선택 

````python
# skiprows 
# n : 지울 행 개수 (상단에서 부터)

excel_rows = pd.read_excel('rawData_sheetplus.xlsx', header=None, skiprows=3)
excel_rows
````

````python
# skiprows=[ ]
# n 번째 값 제외 
excel_rows = pd.read_excel('rawData_sheetplus.xlsx', header=None, skiprows=[3])
excel_rows
````

````python
# skipfooter 
# 뒤에서 부터 제거

excel_rows = pd.read_excel('rawData_sheetplus.xlsx', header=None, skipfooter=3)
excel_rows
````
