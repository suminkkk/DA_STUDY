#### 해변 데이터 > 검색량 이상치 판별 

```python
# 사분위값 
# q1,q2,q3  * 1.5  양끝값 초과시 이상치  -> (현재 데이터에서는)검색량이 너무 많다 -> 혼잡도가 높다 

plt.boxplot(nature['검색건수'])
plt.show()
```

```python
# 상위 경계값 -> 혼잡도가 높을것
# q3 + (q3-q1)*1.5
# 53093.000000 + (53093.000000 -6737.000000) * 1.5
upper_IQR = 53093.000000 + (53093.000000 -6737.000000) * 1.5
upper_IQR
```

```python
import numpy as np
```

```python
# 상위 경계값 초과하는 값(=인기 많아 혼잡할것) 제거 

nature['혼잡도'] = np.where(nature['검색건수']>upper_IQR , '혼잡', '보통')
```
