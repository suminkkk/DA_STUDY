```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model

#tf.keras(tensoflow.keras)는 Tensorflow의 하이레벨을 구현하기위한 Keras API
#tf.keras 모듈의 Sequtial 클래스는 Neural Network의 각 층을 순서대로 쌓는다. : 구조
#ts.keras.layers 모듈의 Dense 클래스는 (완전 연결된: FC)하나의 뉴런층을 구현
#units 뉴런 또는 출력 노드의 개수(양의 정수)
#input_shape는 입력데이터 형태
```

```python
#딥러닝의 실행환경 설정
model.compile(loss="mean_squared_error", optimizer="sgd")
```

```python
#NN 훈련(기계학습)
x = np.array([-1.0,0.0,1.0,2.0,3.0,4.0], dtype=float)
print(x.dtype)

y = np.array([-3.0,-1.0,1.0,3.0,5.0,7.0], dtype=float)
print(y)

#기계학습
#x: train,  y: target(lable)
model.fit(x, y, epochs=500)
```

```python
#예측: x=> 5.0 --->y:?
y_pred = model.predict([5.0])
print(y_pred)
```

```python
model = keras.Sequential([keras.layers.Dense(units=3, input_shape=[1])])
model.compile(loss="mse", optimizer="SGD") #Stochastic Gradient Descent(확률적 경사하강법)
model.fit([1],[[0,1,0]], epochs=1)
```

```python
model.evaluate([1],[[0,1,0]])
#fit 메서드는 훈련진행 상황과 현재의 손실 값을 반환
#evaluate 손실 값을 확인해 보면 손실 값이 감소했다.
```

```python
hist = model.fit([1],[[0,1,0]], epochs=100)
```

```python
print(hist)
print(hist.history)
print(hist.history['loss'])
```

```python
import matplotlib.pyplot as plt  #시각화

loss = hist.history['loss']
#print(loss,":" ,type(loss))

plt.figure(figsize=(4,3))

plt.plot(loss) #y
plt.show()
```

```python
tf.random.set_seed(0)
model = keras.Sequential([keras.layers.Dense(units=3, input_shape=[1])])
model

tf.random.set_seed(0)
#tf.keras.models.clone_model
model2 = tf.keras.models.clone_model(model)
model2
tf.random.set_seed(0)
#tf.keras.models.clone_model
model3= tf.keras.models.clone_model(model)
model3
```

```python
model.compile(loss="mse", optimizer="SGD")
model2.compile(loss="mse", optimizer="Adam")
model3.compile(loss="mse", optimizer="RMSprop")
```

```python
history1 = model.fit([1], [[0,1,0]], epochs = 100, verbose=0)
history2 = model2.fit([1], [[0,1,0]], epochs = 100, verbose=0)
history3 = model3.fit([1], [[0,1,0]], epochs = 100, verbose=0)
```

```python
loss1 = history1.history["loss"]
loss2 = history2.history["loss"]
loss3 = history3.history["loss"]
```

```python
plt.figure(figsize=(4,3))
plt.plot(loss1, label="SGD")
plt.plot(loss2, label="Adam")
plt.plot(loss3, label="RMSprop")

plt.legend()
plt.show()
```

```python
#1. 훈련데이터 준비

x_train = [[0,0],[0,1],[1,0],[1,1]]  #X 2차원 배열 구조
y_train = [[0],[0],[0],[1]]

#2. NN 생성
model = keras.Sequential([
    keras.layers.Dense(units=3,input_shape=[2] , activation='relu'),
    keras.layers.Dense(units=1)
    ])

model.summary()
```

```python
#3.모델 컴파일: 딥러닝 환경설정
model.compile(loss="mse", optimizer="Adam")

#4. NN 훈련: 기계학습

y_pred1 = model.predict(x_train)
print("학습 전: ", y_pred1)

history = model.fit(x_train, y_train, epochs=1000, verbose=0)

y_pred2 = model.predict(x_train)
print("학습 후: ", y_pred2)
```

```python
plt.figure(figsize=(4,3))
loss = history.history["loss"]

plt.plot(loss)

plt.xlabel("epoch")
plt.ylabel("loss")

plt.show()
```
